```python
import numpy as np

A = np.array([
    [10, 20, 30],
    [40, 50, 60],
    [34, 56, 23],
    [70, 80, 90]
])

print(A[[0,1,2]])
print("*"*20)
print(A[:,[1,2]])
print("*"*20)
print(A[[1,2],[1,2]])
print("*"*20)
print(A[A>20])
print("*"*20)

r = [0,1,2]
c = [2,1,0]
print(A[r,c])
```
